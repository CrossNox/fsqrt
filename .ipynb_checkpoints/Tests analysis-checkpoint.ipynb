```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
timeDf = pd.read_csv("time_tests.csv")
fig, ax = plt.subplots(figsize=(12,12))
ax.set_title('Cumsum of time over 10000 random iterations')
line1, = ax.plot(timeDf.index, timeDf.sqrtf.cumsum(), '-', linewidth=1,label='sqrtf')
line2, = ax.plot(timeDf.index, timeDf.fsqrt.cumsum(), '-', linewidth=1,label='fsqrt')
ax.legend(loc='right')
plt.show()
```

```python
piDf = pd.read_csv("pi.csv")
fig, ax = plt.subplots(figsize=(12,12))
ax.set_title('Error on pi calculation')
line1, = ax.plot(piDf.iterations, piDf.difference, '-', linewidth=1)
ax.set_xlabel('Iterations')
ax.set_ylabel('error')
plt.show()
```

```python
diffDf = pd.read_csv("difference_tests.csv")
fig, ax = plt.subplots(4,sharex=True,sharey=True,figsize=(16,24))
#fig.set_title('Difference of fsqrt and sqrtf')
ax[0].semilogx(diffDf.f, diffDf.diff2, linewidth=1,label='2 iterations',basex=2)
ax[1].semilogx(diffDf.f, diffDf.diff3, linewidth=1,label='3 iterations',basex=2)
ax[2].semilogx(diffDf.f, diffDf.diff4, linewidth=1,label='4 iterations',basex=2)
ax[3].semilogx(diffDf.f, diffDf.diff5, linewidth=1,label='5 iterations',basex=2)
ax[3].set_xlabel('value of f')
ax[0].set_ylabel('difference')
ax[1].set_ylabel('difference')
ax[2].set_ylabel('difference')
ax[3].set_ylabel('difference')
ax[0].set_title('2 iterations')
ax[1].set_title('3 iterations')
ax[2].set_title('4 iterations')
ax[3].set_title('5 iterations')
plt.show()
```
